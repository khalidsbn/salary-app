```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/survey_results_public.csv')
```

```python
df = df[['Country', 'EdLevel', 'YearsCodePro', 'Employment', 'ConvertedCompYearly']]

df = df.rename({'ConvertedCompYearly': 'Salary'}, axis=1)
```

```python
df
```

```python
data = df['EdLevel']
data = data.replace({'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor’s degree',
                    'Some college/university study without earning a degree': 'Some college/university',
                    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master’s degree', 
                    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional degree',
                    'Associate degree (A.A., A.S., etc.)': 'Associate degree',
                    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary school'})

data.value_counts().sort_values().plot(kind='barh')
plt.xlabel('Counts')
plt.ylabel('')
plt.title('Degree\'s types');
```

```python
data = df['YearsCodePro']
data = data.dropna()
data = data.replace({'Less than 1 year': 0.5,
                    'More than 50 years': 50})

sns.distplot(data)
plt.xlabel('Years')
plt.title('Distrebution of Experience in years');
```

```python
data = df['Country']
data = data.replace({'United States of America': 'United States',
                    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom'})

data.value_counts()[:15].plot(kind='pie');
plt.ylabel('');
```
